```python
def get_data(text):
    lines = text.split('\n')
    data = []
    for line in lines[:-1]:
        line = line.split()
        episode = int(line[1])
        steps = int(line[3])
        reward = float(line[5][:-1])
        epsilon = float(line[-1])
        data.append([episode, steps, reward, epsilon])
    data = np.array(data)
    return data
```
